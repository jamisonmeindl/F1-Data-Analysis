```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
```

```python
model = joblib.load('model.pkl')
```

```python
pred_data = pd.read_pickle('pred_data.pkl')
```

```python
train_X = pd.read_pickle('train_X.pkl')
```

```python
for col in train_X.columns:
    if col not in pred_data.columns:
        pred_data[col] = 0
```

```python
X_cols = train_X.columns
```

```python
pred_X = pred_data[X_cols]
```

```python
pred_X = pred_X.fillna(pred_X.mean())
```

```python
pred_Y = model.predict(pred_X)
```

```python
results = pd.DataFrame(data = {'DriverNumber': pred_data['DriverNumber'], 'Pred': pred_Y}).sort_values('Pred')
results['PredPosition'] = range(1, len(results) + 1)
```
