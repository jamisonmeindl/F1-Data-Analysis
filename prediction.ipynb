```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
```

```python
model = joblib.load('model.pkl')
```

```python
pred_data = pd.read_pickle('pred_data.pkl')
```

```python
data = pd.read_pickle('train_data.pkl')
```

```python
for col in data.columns:
    if col not in pred_data.columns:
        pred_data[col] = 0
```

```python
pred_X = pred_data[X_cols]
```
