```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
```

```python
model = joblib.load('model.pkl')
```

```python
pred_data = pd.read_pickle('pred_data.pkl')
```

```python
train_X = pd.read_pickle('train_X.pkl')
```

```python
X_cols = train_X.columns
```

```python
pred_data
```

```python
pred_data[X_cols]
```

```python
pred_X = pred_data.groupby('Team')[X_cols].transform(lambda x: x.fillna(x.mean()))
```

```python
pred_X
```

```python
pred_X = pred_X.fillna(pred_X.mean())
```

```python
for col in train_X.columns:
    percentiles = train_X[col].quantile([0, 0.95]).values
    pred_X.loc[pred_X[col] >= percentiles[1], col] = percentiles[1]
```

```python
# pred_X = (pred_X - train_X.mean()) / train_X.std()
```

```python
pred_X
```

```python
pred_Y = model.predict(pred_X)
```

```python
pred_data
```

```python
results = pd.DataFrame(data = {'DriverNumber': pred_data['DriverNumber'], 'Driver': pred_data['Driver'], 'Pred': pred_Y}).sort_values('Pred')
results['PredPosition'] = range(1, len(results) + 1)
```

```python
results
```

```python
results.to_csv('monza_2023')
```
